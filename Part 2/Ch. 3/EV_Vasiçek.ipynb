```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
theta = 0.05  # Long-term mean of the log rate
a = 0.1       # Speed of mean reversion
sigma = 0.02  # Volatility
r0 = 0.03     # Initial short rate
T = 10        # Time horizon
dt = 0.01     # Time step
N = int(T / dt)  # Number of time steps

# Initialize the array to store the short rates
r = np.zeros(N)
r[0] = r0

# Random numbers for the Brownian motion
np.random.seed(42)
dW = np.sqrt(dt) * np.random.randn(N)

# Euler's method to simulate the short rate
for t in range(1, N):
    dr = r[t-1] * (theta + (sigma ** 2) / 2 - a * np.log(r[t-1])) * dt + sigma * r[t-1] * dW[t]
    r[t] = r[t-1] + dr

# Plot the simulated short rate
plt.figure(figsize=(10, 6))
plt.plot(np.linspace(0, T, N), r)
plt.title('Exponential-Vasicek Model')
plt.xlabel('Time [years]')
plt.ylabel('Short rate')
plt.grid(True)
plt.show()
```
